# Data Merging

```python
data1 = {'CustomerID' : [1,2,3,4,5,6], 'Product' : ['A', 'B','C', 'A', 'C','F']}
data1
```

```python
data2 =  {'CustomerID' : [2,4,6,7], 'State' : ['S1', 'S2','S3', 'S4']}
data2
```

```python
dfSales = pd.DataFrame(data1)
dfSales
```

```python
dfRegion = pd.DataFrame(data2)
dfRegion
```

```python
#Inner Join
pd.merge(dfSales, dfRegion, how= 'inner', on='CustomerID')
```

```python
#Left Join
pd.merge(dfSales, dfRegion, how= 'left', on='CustomerID')
```

```python
#right Join
pd.merge(dfSales, dfRegion, how= 'right', on='CustomerID')
```

```python
#Outer Join
pd.merge(dfSales, dfRegion, how= 'outer', on='CustomerID')
```
